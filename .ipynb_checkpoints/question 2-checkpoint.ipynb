```python
#if, elif, else: Used for conditional statements to execute different blocks of code based on specified conditions.
x = 10

if x > 5:
    print("x is greater than 5")
elif x == 5:
    print("x is equal to 5")
else:
     print("x is less than 5")
```

```python
#for, while: Used for looping constructs to iterate over sequences (like lists, tuples, or strings) or execute a block of code repeatedly while a condition is true.


# Example using for loop
numbers = [1, 2, 3, 4, 5]

for num in numbers:
    print(num)

# Example using while loop
count = 0
while count < 5:
    print(count)
    count += 1
```
